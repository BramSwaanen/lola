# Evaluate wrongly/correctly classified

#### Load evaluations

```python
import pandas as pd
from evaluation import get_wrong_items, get_fol_expressions, get_free_vars, get_correct_items
```

```python
#Load csv with results 
res = pd.read_csv('../evaluations/sick_train_evaluation.csv', sep = '\t')
res
```

#### Get wrongly classified

```python
wrong_e, wrong_c, wrong_n = get_wrong_items(res)   
wrong = pd.concat([wrong_e, wrong_c, wrong_n])
```

```python
#all wrongly classified items 
wrong
```

#### Get correctly classified

```python
correct_e, correct_c, correct_n = get_correct_items(res)
correct = pd.concat([correct_e, correct_c, correct_n])
```

```python
correct
```

#### Ananlyse number of free variables in both wrongly and correctly classified

```python
#Get all FOL expressions 
wrong_fol_expr = get_fol_expressions(wrong) 
print(len(wrong_fol_expr))
correct_fol_expr = get_fol_expressions(correct)
print(len(correct_fol_expr))
```

```python
wrong_free_vars = get_free_vars(wrong)
correct_free_vars = get_free_vars(correct)
```

```python
#instances with free variables in wrongly classified with their predicted and gold label
wrong_free_vars
```

```python
#Instances with free variables in correctly classifed with their predicted and gold label
correct_free_vars
```
